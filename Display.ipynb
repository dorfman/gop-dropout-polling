Show how polling changes when a Republican candidate suspends their campaign for the presidential nomination. Only examined candidates that dropped out during primary season.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import numpy as np
import pandas as pd
import os
from IPython.display import display, HTML
import stats
import analysis
```

## Read Written Files

### bootPolls.csv
Downloaded as a CSV from [Huffington Post's Pollster](http://elections.huffingtonpost.com/pollster/2016-national-gop-primary). More info on cleaning polling data in [02-DataCleaning.ipynb](https://data301a.calpolydatascience.org/user/dorfman/notebooks/project/02-DataCleaning.ipynb) and [03-Exploration.ipynb](https://data301a.calpolydatascience.org/user/dorfman/notebooks/project/03-Exploration.ipynb).

### candidates.csv
Manually made dictionary using data (from [Wikipedia](https://en.wikipedia.org/wiki/United_States_presidential_election,_2016#Withdrawn_candidates_2
)) on when candidates dropped out.

### PollingBeforeDrop.csv and PollingAfterDrop.csv
Create two DataFrames indexed by the candidates that dropped out of the race during the primaries. Each column in PollingBeforeDrop holds other candidates' average polling numbers in the week prior to a candidate's campaign suspension. Each cell contains the polling for a candidate in the column the week before the candidate in the row dropped out. Each column in PollingAfterDrop holds other candidates' average polling numbers in the week after a candidate suspends their campaign. Each cell contains the polling for a candidate in the column the week after the candidate in the row dropped out. The average is weighted heavier towards days further away from the date of suspension. More info on these DataFrames in [04-Modeling.ipynb](https://data301a.calpolydatascience.org/user/dorfman/notebooks/project/04-Modeling.ipynb).

```python
polls = analysis.CleanPolls()
candidates = analysis.GetDropData()
PollingBeforeDrop = analysis.GetBeforeData()
PollingAfterDrop = analysis.GetAfterData()
PollingDiff = PollingAfterDrop - PollingBeforeDrop
statFrame = analysis.GetStats(PollingBeforeDrop, PollingAfterDrop, PollingDiff)
pollList = analysis.GenPollSnippets(polls, candidates)
```
